# UNIFICACION DE ARCHIVOS

```python
import pandas as pd
```

```python
#lectura de archivos y almacenamiento en variables data frames

df_customers = pd.read_csv('../datos_originales/olist_customers_dataset.csv')
df_geolocation = pd.read_csv('../datos_originales/olist_geolocation_dataset.csv')
df_order_items = pd.read_csv('../datos_originales/olist_order_items_dataset.csv')
df_order_payments = pd.read_csv('../datos_originales/olist_order_payments_dataset.csv')
df_order_reviews = pd.read_csv('../datos_originales/olist_order_reviews_dataset.csv')
df_orders = pd.read_csv('../datos_originales/olist_orders_dataset.csv')
df_products = pd.read_csv('../datos_originales/olist_products_dataset.csv')
df_sellers = pd.read_csv('../datos_originales/olist_sellers_dataset.csv')
df_category_translation = pd.read_csv('../datos_originales/product_category_name_translation.csv')
```

```python
# unificacion de data frames

df_unificado = df_orders.merge(df_order_items, on='order_id', how='left') \
                        .merge(df_order_payments, on='order_id', how='left') \
                        .merge(df_order_reviews, on='order_id', how='left') \
                        .merge(df_customers, on='customer_id', how='left') \
                        .merge(df_products, on='product_id', how='left') \
                        .merge(df_sellers, on='seller_id', how='left') \
                        .merge(df_category_translation, on='product_category_name', how='left')
```

```python
#guardar data frame unificado en nuevo csv unico
df_unificado.to_csv("../datos/olist_unificado.csv", index=False)
```


* En este notebook se llevó a cabo la unificación de 9 archivos CSV independientes del dataset en un único DataFrame consolidado. Este proceso es   
  fundamental para poder realizar análisis cruzados que relacionen información de pedidos, clientes, productos, vendedores, pagos y reseñas.

* La estrategia de unificación utilizó múltiples merges secuenciales con tipo 'left' para preservar todos los registros de la tabla principal (orders) y 
  añadir información relacionada de las demás tablas:

* El DataFrame unificado resultante contiene 119,143 registros con 40 columnas, listo para el análisis exploratorio.
